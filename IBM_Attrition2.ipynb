```python
#removing meaningless columns
df=df.drop('EmployeeCount',axis=1)
df=df.drop('StandardHours',axis=1)
```

```python
df.corr()
```

```python
#transforming non-numerical to numerical col
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column in df.columns:
    if df[column].dtype== np.number:
        continue
    df[column]=le.fit_transform(df[column])
```

# A breakdown of distance from home by job role and attrition.


```python
distancefrhome_JobRole_attrition = df[["DistanceFromHome", "Attrition", "JobRole"]]
distancefrhome_JobRole_attrition=distancefrhome_JobRole_attrition .replace({"JobROLE": {1:'Healthcare Representative',2:'Human Resources',3:'Laboratory Technician',4:'Manager',5:'Manufacturing Director',6:'Research Director',7:'Research Scientist',8:'Sales Executive',9:'Sales Representative'}})
distancefrhome_JobRole_attrition= distancefrhome_JobRole_attrition.groupby(["JobRole", "Attrition"]).agg('mean')
print(distancefrhome_JobRole_attrition)
distancefrhome_JobRole_attrition.unstack(level=-1)['DistanceFromHome'].plot(kind='bar')
```

# average monthly income by education and attrition


```python
edu_attrition_income = df[["Education", "Attrition", "MonthlyIncome"]]
edu_attrition_income = edu_attrition_income.groupby(["Education", "Attrition"]).agg('mean')
print(edu_attrition_income)
edu_attrition_income.unstack(level=-1)['MonthlyIncome'].plot(kind='bar')
```
